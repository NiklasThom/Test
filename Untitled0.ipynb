```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage to scrape
url = 'https://www.fupa.net/player/stephan-kumpfmueller-552043'

# Send a GET request to the webpage
response = requests.get(url)
response.raise_for_status()  # Check for request errors

# Parse the webpage content
soup = BeautifulSoup(response.text, 'html.parser')

# Print the parsed HTML (for inspection)
print(soup.prettify())

```

```python
name_tag = soup.find('h1', class_='Spiele')  # Replace 'player-name' with the actual class
player_name = name_tag.get_text(strip=True) if name_tag else 'N/A'

# Extract player position
position_tag = soup.find('div', class_='player-position')  # Replace 'player-position' with the actual class
player_position = position_tag.get_text(strip=True) if position_tag else 'N/A'

# Print extracted data
print(f"Player Name: {player_name}")
print(f"Position: {player_position}")

```
